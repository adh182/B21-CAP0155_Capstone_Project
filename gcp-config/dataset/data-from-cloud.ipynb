## GET TRAINING AND TESTING DATA FROM GOOGLE CLOUD STORAGE BUCKET

The storage URL is public to the internet. Do not share the URL!

Do not edit this notebook. Only copy the code to your programs.

```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd

def get_data(url):
    '''Convert data fetched from API into pandas dataframe'''
    
    #API requests
    r = requests.get(url)
    #Parsing XML
    root = ET.fromstring(r.content)
    
    file_name = []
    for child in root:
        for item in child:
            if 'earthquake-damage-dataset' in item.text:
                file_name.append(item.text)
                
    testing_data = pd.read_csv(url+'/'+file_name[0])
    training_labels = pd.read_csv(url+'/'+file_name[1])
    training_data = pd.read_csv(url+'/'+file_name[2])
    
    return training_data, training_labels, testing_data
```

```python
url = 'http://storage.googleapis.com/b21-cap0155-capstone-project-bucket-1'

training_data, training_labels, testing_data = get_data(url)
```
